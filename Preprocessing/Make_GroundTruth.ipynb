```python
import SimpleITK as sitk
import numpy as np
import os
import os.path as osp
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib
```

## Ground Truth Estimation

### Majority Voting
Use majority voting to obtain the reference segmentation. Note that this filter does not resolve ties. In case of ties it will assign the backgtound label (0) to the result.

```python
#export
def m_voting(segmentations, labelForUndecidedPixels = 0):
    'Majority Voting from  simple ITK Label Voting'
    segmentations = [sitk.GetImageFromArray(x) for x in segmentations]
    mv_segmentation = sitk.LabelVoting(segmentations, labelForUndecidedPixels)
    return sitk.GetArrayFromImage(mv_segmentation)
```

### GroundTruth Estimator

```python
inputdirs = '/content/drive/MyDrive/Colab/Mask_experts'
outputdir = '/content/drive/MyDrive/MyProject/Mask'
```

```python
experts_annotation = ['Maps1_T', 'Maps2_T', 'Maps3_T', 'Maps4_T', 'Maps5_T', 'Maps6_T']
```

```python
def Ground_Truth_Estimator(item, inputdirs, experts_annotation, outputdir):
    imgs = []
    for p in experts_annotation:
        p = f'{inputdirs}/{p}/{item}'
        if osp.isfile(p):
            annotation = np.array(Image.open(p))
            annotation[annotation == 6] = 1
            imgs.append(annotation)
    if len(imgs) > 0:
        multilabel_STAPLE = m_voting(imgs)
        result = Image.fromarray(multilabel_STAPLE)
        result.save(f'{outputdir}/{item}')
    else:
        print('not annotation !')
```

```python
train_dir = '/content/drive/MyDrive/Colab/Train_dataset'
for img in os.listdir(train_dir):
    item = img.replace('.jpg', '_classimg_nonconvex.png')
    print(item)
    Ground_Truth_Estimator(item, inputdirs, experts_annotation, outputdir)
```
